## 이미지 출력 기본

```python
%matplotlib inline
from matplotlib.pyplot import imread,imshow,axis
img=imread('f.png')
# Color

plt.axis('off')


imshow(img)
```

## 두번재 체널을 출력(y축 보임)

```python
fig=plt.imshow(img[:,:,1])
ax=plt.gca()
ax.get_xaxis().set_visible(False)
```

## 평균을 취해서 Gray로 출력

```python
import numpy as np
plt.axis('off')
imshow(np.mean(img,axis=2), cmap=plt.cm.gray)
```
