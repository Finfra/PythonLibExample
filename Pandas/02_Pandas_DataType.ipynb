## Series

Index 와 Value 로 구성되어 Python 의 Dictionary 와 유사한 형태이다.

### 데이터 생성

#### List 로 생성하기

```
data = pd.Series([1,2,3,4])
data
```

#### Tuple 로 생성하기

```
data = pd.Series((1,2,3,4))
data
```

#### Array 로 생성하기

```
import numpy as np
data = pd.Series(np.array([1,2,3,4]))
data
```

#### Dictionary 로 생성하기

```
data = pd.Series({0 : 1, 1 : 2, 2 : 3, 3 : 4.5})
data
```

#### 여러 자료 형 index 를 넣어 생성하기

```
data = pd.Series(["abc", 3, 4.5], index=["initial","ea","float"])
data
```

### 데이터 접근

#### Index(키) 값 가져오기

```
data.index
```

#### Value 값 가져오기

```
data.values
```

#### 특정 값 접근

```
data[2]
```

#### Index 명 바꾸기

```
data.index = ["what", "when", "where"]
data
```

#### 조회하기

##### index 로 조회하기

```
data.at["what"]
```

```
data[["what","where"]]
```

##### 범위로 조회하기

```
data[0:2]
```

```
data[-2:]
```
